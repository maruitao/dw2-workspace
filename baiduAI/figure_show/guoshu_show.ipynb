```python
import pandas as pd
import numpy as np
import openpyxl
pd.set_option('display.max_colwidth', 200)
```

```python
def get_1st_item(s):
    str_corr = s[36:44]
    #  去掉下面不需要字符
    result = str_corr.strip('u},\'\'').replace(',','').replace('}','').replace('{','').replace('\'','').replace('u','').replace(' ','')
    
    return result

def read_file(filepath):
    df = pd.read_csv(filepath,encoding='utf-8')
    df.drop(df.columns[[0,1,3]], inplace=True, axis=1) # inplace=True 表示在本df修改 不需要再赋值
    return df
"""
jmc-76s.csv  jmc-result-508s.csv
qc-73s.csv  qc-result-492s.csv
qincai-74s.csv   qincai-result-525s.csv
"""
df = pd.read_csv('dataset/jmc-76s.csv', encoding='utf-8')

# df.drop(df.columns[[0,1,3]], inplace=True, axis=1) # inplace=True 表示在本df修改 不需要再赋值
df
```

```python
def get_df_by_series(series):
    dataframe  =  pd.DataFrame(data=series, columns=['result'])
    return dataframe
    
```
