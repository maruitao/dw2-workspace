```python
# -*- coding:utf-8 -*-
import pandas as pd
import jieba
from wordcloud import WordCloud
import sys 
# reload(sys) 
# sys.setdefaultencoding("utf-8")
```

```python
df = pd.read_csv('all_data.csv', encoding='utf-8')
df.head
```

```python
def read_stopwords():
    stopwds = []
    with open('chn-stop-words.txt','r') as fh:
        for w in fh :
            stopwds.append(w.strip())
    fh.close()
    return stopwds
```

```python
stopwds = read_stopwords()
word_list = []
temp = None
```

```python
for val in df['title']:
    temp = jieba.cut(str(val))
    temp = filter(lambda x : len(x) > 1 and x not in stopwds , temp)
    for i in temp :
        if len(i) > 1:
            word_list.append(i.encode('utf-8'))
result = " ".join(word_list)
print type(result)
```

```python
import matplotlib.pyplot as plt
# from scipy.misc import imread
from PIL import Image  
import numpy as np  
# color_mask = imread('bigdata.jpg')
color_mask =  np.array(Image.open("bigdata.jpg"))  
# wordcloud = WordCloud(font_path='simsun.ttf',background_color="white", max_words=500, margin=1, mask=color_mask).generate(result.decode())
wordcloud = WordCloud(font_path='simsun.ttf',background_color="white", max_words=500, margin=1).generate(result.decode())
plt.imshow(wordcloud, interpolation = 'bilinear')

plt.axis('off')
# plt.imshow(alice_coloring, cmap=plt.cm.gray)
plt.show() # 目前词云有英文和中文,因为在调用时传入unicode对象
wordcloud.to_file("title_wordcloud.png")
```
